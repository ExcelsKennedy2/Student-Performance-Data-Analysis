```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['g3'].describe()
```

```python
# Performance by study time
# Does studying more actually improve grades?

df.groupby('studytime')['g3'].mean().sort_values(ascending=False)
```

```python
# Impact of absences

df.groupby(pd.cut(df['absences'], bins=[-1, 5, 10, 20, 100]))['g3'].mean()
```

```python
# family support & grades
# Does family or school support make a difference?

df.groupby('famsup')['g3'].mean()
```

```python
df.groupby('schoolsup')['g3'].mean()
```

```python
# Parental Education Impact

df.groupby('medu')['g3'].mean()
```

```python
df.groupby('fedu')['g3'].mean()
```

```python
# Alcohol consumption VS Grades

df['alcohol_score'] = df['dalc'] + df['walc']
```

```python
# Does alcohol use affect academic performance?

df.groupby('alcohol_score')['g3'].mean().sort_values()
```

```python
# Identify at-risk students

at_risk = df[
    (df['g3'] < 10) & (df['absences'] > 10)
]
at_risk.head()
```

```python
# creating performance categories

df['performance_level'] = pd.cut(
    df['g3'],
    bins=[-1, 9, 13, 20],
    labels=['Low', 'Average', 'High']
)
```

```python
df['performance_level'].value_counts()
```

```python
# Correlation Check 
# Show which factors matter most

df[['g1', 'g2', 'g3', 'studytime', 'absences', 'failures']].corr()
```

```python
df.to_csv("cleaned_student_performance.csv", index=False)
```

```python
# Distribution of Final Grades (G3)
# How are students generally performing?

df['g3'].plot(kind='hist', bins=20, title='Distribution of Final Grades')
```

```python
# Study time vs Final Grade
# Do students who study more perform better?

df.boxplot(column='g3', by='studytime')
```

```python
# Absences VS Grades
# Do absences affect grades?

df.plot(
    x='absences',
    y='g3',
    kind='scatter',
    title='Absences vs Final Grade'
)
```

```python
# Family Support Comparison
# Does family support make a difference?

df.groupby('famsup')['g3'].mean().plot(kind='bar', title='Family Support vs Final Grade')
```

```python
# Alcohol Consumption vs Performance
# Does alcohol consumption affect academic results?

df.boxplot(column='g3', by='alcohol_score')
```

```python
# Performance Levels (Low/Average/High)

df['performance_level'].value_counts().plot(
    kind='bar',
    title='Student Performance Levels'
)
```
